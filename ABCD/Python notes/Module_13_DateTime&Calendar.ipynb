The way date and time is represented may be different in different places, organizations etc. It's more common to use **mm/dd/yyyy** in the US, whereas **dd/mm/yyyy** is more common in the UK.

Python has **strftime()** and **strptime()** methods to handle this.

## Datetimes Directives

# strftime()
- **datetime object to string** <br>
- The strftime() method is defined under classes date, datetime and time. The method creates a formatted string from a given date, datetime or time object.

```python
print(x.strftime("%A"))   # %A is a directive which specifies the day of the week
```

```python
print(x.strftime("%Y"))
```

```python
print(x.strftime("%B"))
```

```python
print(x.strftime("%M"))
```

```python
print(x.strftime("%Y,%A,%M,%H,%I"))
```

## strptime()
- **string to datetime**

### Different ways of formatting a string date into a datetime format:

```python
rawDate = "11-September 2021"
a = datetime.strptime(rawDate,'%d-%B %Y')
print(a)
```

```python
from datetime import datetime

date_string0 = "21Jun201822234556"
date_string1 = "21 Jun 2018 22 234556"
date_string2 = "21,June-2018:22.234556"


print("date_string0 before formatting =", date_string0)
date_object0 = datetime.strptime(date_string0, "%d%b%Y%S%f")
print("date_object0 after formatting =", date_object0)
print()


print("date_string1 before formatting =", date_string1)
date_object1 = datetime.strptime(date_string1, "%d %b %Y %S %f")
print("date_object1 after formatting =", date_object1)
print()

print("date_string2 before formatting =", date_string2)
date_object2 = datetime.strptime(date_string2, "%d,%B-%Y:%S.%f")
print("date_object2 after formatting =", date_object2)
print()
```

## How to Build your own clock ?

```python
import time

while True:
    localtime = time.localtime()
    clock = time.strftime('%I:%M:%S:%p', localtime)
    print(clock, end='\r')
#     print('\r', end = "")
    time.sleep(0.5)
```

# NOTE

**print()** adds a \n at the end of what is being printed. This can be changed with the end parameter. Output from print() goes into a buffer. 

When you change the end parameter, the buffer no longer gets flushed. **To ensure that you get output as soon as print() is called, you also need to use the flush=True parameter**

### EXAMPLE:


```python
import time

def count_items(items):
    print('Counting ', end='')
    num = 0
    for item in items:
        num += 1
        time.sleep(1)
        print('.', end='')

    print(f'\nThere were {num} items')

count_items("hello")
```

# Calendar

```python
import calendar    # import the calendar module to use the calendar feature
```

## Year Calendar

```python
# initialize the year
year = 2021

# printing the calendar
print(calendar.calendar(year))
```

## Month Calendar

```python
# we have already imported the calendar module

# initializing the year and month
year = 2021
month = 9
# printing the calendar
print(calendar.month(year, month))
```

## Find the number of leap years

```python
print("The number of leap days between 1980 and 2021 are:")

print(calendar.leapdays(1980, 2021))
```
