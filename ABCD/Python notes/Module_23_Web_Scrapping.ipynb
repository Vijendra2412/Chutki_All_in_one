## IMDB TOP 100 movie data extraction using WEB SCRAPING

**urllib.request** is a python module for fetching URLs (Uniform Resource Locator). It offers a very simple interface in the form of the **urlopen** function.<br>
**urlopen** fucntion is capable of fethching URLs using a variety of different protocols.<br>

**Beautiful Soup** is a library that makes it easy to scrape information from web pages. It **sits atop an HTML or XML parser**, providing Pythonic idioms for iterating , searching, and modifying the parse tree.


```python
# importing the required libraries from python 
from urllib.request import urlopen
from bs4 import BeautifulSoup as soup
```

```python
my_url = "https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc"
uClient = urlopen(my_url)
page_html = uClient.read() #FOR READING THE WEBPAGE DATA
uClient.close()
```

```python
page_soup = soup(page_html,"html.parser")  # PASSING THE PAGE DATA TO BEAUTIFUL SOUP OBJECT
### This module defines a class HTMLParser which serves as the basis for parsing text files formatted in HTMLand XHTML.
s = soup(uClient)
```

**findAll** findAll() returns all the tags and strings that match your filters. This can take a while if the document is large. If you donâ€™t need all the results, you can pass in a number for limit. This works just like the LIMIT keyword in SQL.

<br>Example: **soup.findAll('a', limit = 2)**
**soup('a')** is similar to above. <br>
If you treat the BeautifulSoup object or a Tag object as though it were a function

```python
containers = page_soup.findAll('div',{'class':'lister-item mode-advanced'})
#findAll() is the most popular
print(len(containers))
```

We can now print out the **HTML content of the page, formatted nicely, using the prettify method** on the BeautifulSoup object. This step isn't strictly necessary, and we won't always bother with it, but it can be helpful to look at prettified HTML to make the structure of the and where tags are nested easier to see

```python
print(soup.prettify(containers[2]))    #to farmat the HTMl page nicely
```

```python
filename = "imdb_top_100.csv"     #create a csv file named imdb_top_100.csv and store it in a variable named filename
```

```python
f = open(filename, "w")   #w s for writing
```

```python
headers = "Name, Year, Runtime\n"
f.write(headers)
```

```python
for container in containers:    #iterate over all the 50 values stored in containers and store in container
    name = container.img["alt"]    #find the 'img' tag on IMDB website(using the above url) and store 'alt' value in name
    
    year_mov = container.findAll("span",{"class":"lister-item-year text-muted unbold"})
    year = year_mov[0].text
    
    runtime_mov = container.findAll("span",{"class":"runtime"})
    runtime = runtime_mov[0].text
    
    print(name + "," + year + "," + runtime + "\n")
    f.write(name + "," + year + "," + runtime + "\n")
f.close()
```
