```python
from bs4 import BeautifulSoup
import requests
```

```python
url ='https://www.flipkart.com/search?q=laptop&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=off&as=off&as-pos=1&as-type=HISTORY&page='
```

```python
e = []
for i in range(2,6):
    link = url + str(i)
    read = requests.get(link)
    content = read.content
    soup = BeautifulSoup(content,'html.parser')
    details = soup.find_all('li',{'class':'rgWa7D'})
    for i in range(len(details)):
        e.append(details[i].text)
    
```
