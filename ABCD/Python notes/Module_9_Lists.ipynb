## What are lists ?

- among the **many collection data types**(tuple, set, dictionary) in python, list is one of them.
- used to store multiple elements of different types in a single variable.
- list elements are mutable/changeable 
- list elements are ordered
- list allows duplicates values
- to create lists use [ ] or list()

## How to add items in lists
- insert()
- append()
- extend()


```python
mylist = ['math','physics','chemistry']
mylist.insert(3, 'computer')
print(mylist)
```

```python
print("Before append: ", mylist)
mylist.append('commerce')
print("After append: ", mylist)
```

```python
mylist = ['math','physics','chemistry']
mylist.append(('commerce', 'civil'))
print("After append: ", mylist)
```

```python
mylist = ['math','physics','chemistry']
mylist.extend(('commerce', 'civil'))
print("After extend: ", mylist)
```

```python
mylist = ['math','physics','chemistry']
mylist.extend('civil')
print("After extend: ", mylist)
```

## What is the difference between append() & extend() ?
- **append()** : Adds its argument as a single element to the end of a list. The length of the list increases by one. ... 
- **extend()** : Iterates over its argument and adding each element to the list and extending the list. The length of the list increases by number of elements in it's argument.

## Looping in Lists
- using for loop
- using range() and len() functions
- using while loop

```python
mylist = ['math', 'physics', 'chemistry', 'computer', 'commerce']
```

```python
# using simple for loop to iterate over the list items

for i in mylist:
    print(i)
```

```python
# using range() and len() functions

for i in range(len(mylist)):
    print(mylist[i])
```

```python
# using while loops to iterate over list items

i = 0
while i < len(mylist):
    print(mylist[i])
    i += 1
```

## List Comprehensions
-  List comprehension offers a shorter syntax when you want to create a new list based on the values of an existing list.


```python
mylist = ['math', 'physics', 'chemistry', 'computer', 'commerce']
```

```python
# using list comprehensions to do the same task but within single line

[x for x in mylist]
```

```python
[x for x in mylist if x.startswith('c')]    #filter out the items which starts with 'c'
```

```python
numlist = list(range(1,11))
print("Before: ",numlist)
```

```python
[x for x in numlist if x%2==0]              # using list comprehensions to print the even values
```

```python
mylist = ['math', 'physics', 'chemistry', 'computer', 'commerce']
```

```python
[x for x in mylist if x != "commerce"]
```

```python
[x.upper() for x in mylist]
```

```python
mylist = ['apple', 'banana', 'apricot','lemon']
empty=[]
empty = [x for x in mylist if x.startswith('a')]
print(empty)
```

## Sort() 
- sort() method can be used to sort a list
- case sensitive... Capital letters are given the preference

```python
mylist
```

```python
# sort the list element and store it in the same object
mylist.sort()
mylist
```

```python
# sort in reverse order
mylist.sort(reverse=True)
print(mylist)
```

```python
mylist = ['Lemon', 'Banana', 'apricot', 'apple']
mylist.sort()
print(mylist)                   #capital letters are given the preference over small letters
```

```python
mylist = ['Lemon', 'Banana', 'apricot', 'apple']
mylist.sort(key = str.lower)   

#key parameter can be used to apply the lower method to smallcase the elements before sorting
print(mylist)   
```

```python
# to print the values nearest to 50 in sorted format

def near50(n):
    return abs(n-50)       #abs() will give the absolute value

mylist = [100, 60,50,84,30]
mylist.sort(key = near50)
print(mylist)
```

## copy() lists
- using copy()
- using list()

```python
list1 = ['hello', 'how']
list2 = list1.copy()              # using 'copy()' method to copy the lists into other lists
print("this is copied from list1: ",list2)
```

```python
list1 = ['hello', 'how']
list2 = list(list1)               #using 'list()' method we can copy lists
print("this is copied from list1: ",list2)
```
